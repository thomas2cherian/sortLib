```python
# IMPORT standard functions
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# IMPORT probeinterface tools
from probeinterface import Probe, ProbeGroup, write_probeinterface, write_prb
from probeinterface.plotting import plot_probe, plot_probe_group
```

```python
# CREATE two probes to mimic the two depths of the shanks in FMA
# Each probe will behave like a single multi contact shank
contactRadius = 2.5    
df            = pd.read_csv('cocoFma256.csv')
xy            = df[['x','y']].values

fma256 = Probe(ndim = 2, si_units = 'um')
fma256.set_contacts(positions = xy, shapes = 'circle', shape_params = {'radius': contactRadius})
fma256.set_device_channel_indices(df['hsw'].values)
fma256.set_contact_ids(df['fma'].values)
fma256.set_shank_ids([0]*256)
fma256.create_auto_shape(probe_type = 'tip')

#fma256_3d = fma256.to_3d(axes='xz')
plot_probe(fma256)

#plt.show()

```
